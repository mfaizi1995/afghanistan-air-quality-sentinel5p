# Air Quality Trends in Afghan Cities (2019–2024)
### A Multi-Pollutant Satellite-Based Environmental Analysis

#### Author: Mastoorah Faizi
#### Data: Sentinel-5P TROPOMI (OFFL/L3)
#### Period: January 2019 – December 2024
---

## Abstract

This report analyzes six years of satellite-based air quality data over major Afghan cities using the Sentinel-5P TROPOMI sensor. We focus on two key atmospheric pollutants:

- **NO₂ (nitrogen dioxide)** – traffic, generators, heating
- **CO (carbon monoxide)** – incomplete combustion, trash burning, cooking/heating fires

We compute monthly averages, multi-year trends, seasonal cycles, pollutant correlations, and city-to-city differences. The analysis highlights strong winter pollution peaks, clear seasonal cycles, and meaningful differences between cities.
